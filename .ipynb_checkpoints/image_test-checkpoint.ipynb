```python
# Import libraries
import numpy as np
import pandas as pd
from obspy import read
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import os

import cv2 as cv
```

```python
cat_directory = './data/lunar/training/catalogs/'
cat_file = cat_directory + 'apollo12_catalog_GradeA_final.csv'
cat = pd.read_csv(cat_file)
```

```python
row = cat.iloc[7]
arrival_time = datetime.strptime(row['time_abs(%Y-%m-%dT%H:%M:%S.%f)'],'%Y-%m-%dT%H:%M:%S.%f')

arrival_time_rel = row['time_rel(sestatsc)']

test_filename = row.filename
```

```python
data_directory = './data/lunar/training/data/S12_GradeA/'
mseed_file = f'{data_directory}{test_filename}.mseed'
st = read(mseed_file)
st
```

```python
st[0].stats
```

```python
# This is how you get the data and the time, which is in seconds
tr = st.traces[0].copy()
tr_times = tr.times()
tr_data = tr.data

# Start time of trace (another way to get the relative arrival time using datetime)
starttime = tr.stats.starttime.datetime
arrival = (arrival_time - starttime).total_seconds()
arrival
```

```python
# Set the minimum frequency
minfreq = 0.5
maxfreq = 1.0

# Going to create a separate trace for the filter data
st_filt = st.copy()
st_filt.filter('bandpass',freqmin=minfreq,freqmax=maxfreq)
tr_filt = st_filt.traces[0].copy()
tr_times_filt = tr_filt.times()
tr_data_filt = tr_filt.data
```

```python
# To better see the patterns, we will create a spectrogram using the scipy function
# It requires the sampling rate, which we can get from the miniseed header as shown a few cells above
from scipy import signal
from matplotlib import cm
f, t, sxx = signal.spectrogram(tr_data_filt, tr_filt.stats.sampling_rate)
```

```python
img = cv.imread('my_plot.png', cv.IMREAD_GRAYSCALE)

print(img.shape)

kernel = np.ones((6,6), np.uint8)
dilation = cv.dilate(img, kernel, iterations = 1)

# Display the image using plt.imshow()
plt.imshow(dilation)

# Show the plot
plt.show()

#cv.imshow('image',dilation)
#cv.waitKey(0)
#cv.destroyAllWindows()
```

```python
# Threshold it so it becomes binary
ret, thresh = cv.threshold(dilation,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)

# Use connectedComponents() (returns a labeled image and the number of labels)
num_labels, labels = cv.connectedComponents(thresh)

# Use connectedComponentsWithStats() (returns a labeled image, the number of labels, and statistics for each component)
num_labels, labels, stats, centroids = cv.connectedComponentsWithStats(thresh)
```

```python
#fig = plt.figure(figsize=(15, 10))

# Display the image using plt.imshow()
plt.imshow(labels)

# Show the plot
plt.show()
```

```python
nb_components, output, stats, centroids = cv.connectedComponentsWithStats(thresh, connectivity=4)
sizes = stats[:, -1]

max_label = 2
max_size = sizes[1]
for i in range(2, nb_components):
    if sizes[i] > max_size:
        max_label = i
        max_size = sizes[i]

img2 = np.zeros(thresh.shape)
img2[output == max_label] = 255

x = stats[2, cv.CC_STAT_LEFT]
y = stats[2, cv.CC_STAT_TOP]
w = stats[2, cv.CC_STAT_WIDTH]
h = stats[2, cv.CC_STAT_HEIGHT]

print(x, y)

#cv.rectangle(img2, (x, y), (x + w, y + h), (0, 255, 0), 3)

i = 0
j = x

while i < img2.shape[0]:
    #print(img2[i][j])
    img2[i][j] = np.uint8(255)
    img2[i][j+w] = np.uint8(255)
    #print(img2[i][j])
    i += 1

fig = plt.figure(figsize=(15, 10))

# Display the image using plt.imshow()
plt.imshow(img2)

# Show the plot
plt.show()

#cv.imshow('image',img2)
#cv.waitKey(0)
#cv.destroyAllWindows()
```
